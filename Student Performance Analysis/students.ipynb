```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Student": ["A","B","C","D","E","F","G","H","I","J"],
    "Study_Hours": [2, 3, 5, 1, 4, 6, 2, 7, 5, 3],
    "Sleep_Hours": [6, 7, 5, 8, 6, 5, 7, 4, 6, 7],
    "Marks": [50, 60, 80, 40, 70, 85, 55, 90, 75, 65],
    "Subject": ["Math","Science","Math","English","Science",
                "Math","English","Science","Math","English"]
}

df = pd.DataFrame(data)

max_marks = df["Marks"].max()
top_student = df["Marks"].idxmax()

print(top_student)
```

```python
fig , ax = plt.subplots(2,2,figsize = (12,8))

ax[0,0].plot(df["Student"] , df["Marks"] , color = "red" , marker = "o" , label = "Marks vs Students")
ax[0,0].set_title("Line Plot")
ax[0,0].legend()
ax[0,0].set_xlabel("Students")
ax[0,0].set_ylabel("Marks")
ax[0,0].plot(df["Student"][top_student] , max_marks , color = "black" , marker = "^" , label = "Top Student")
ax[0,0].legend()
```

```python
avg_marks = df.groupby("Subject")["Marks"].mean().reset_index(name = "Avg_marks")

tdf = pd.DataFrame(avg_marks)

ax[0,1].bar(tdf["Subject"] , tdf["Avg_marks"] , color = "blue")
ax[0,1].set_title("Average Marks per Subject")
ax[0,1].set_xlabel("Subjects")
ax[0,1].set_ylabel("Average Marks")

plt.show()
```
